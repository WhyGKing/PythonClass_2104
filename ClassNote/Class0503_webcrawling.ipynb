##### reqeust 모듈
HTTP : request 와 response로 구성

request 메서드: get post put delete
- get // 정보를 가져오기
- post // 새로운정보 내보내기
- put // 정보수정
- delete // 정보삭제

response 응답코드
- 1XX // 요청을 받았고, 작업진행중
- 2XX // 요청 수행성공
- 3XX // 요청 완료, 리다이렉션 필요
- 4XX // 사용자 요청이 잘못됨
- 5XX // 서버오류
