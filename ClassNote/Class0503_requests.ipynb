```python
# 모듈 import
import requests
```

```python
URL = 'https://www.naver.com'
naverhome = requests.get(URL)
print(naverhome.status_code)
```

```python
print(naverhome.text)
```

```python
# 구글검색페이지 이용, python 검색결과 가져오기
URL = 'https://www.google.com/search'
parameters = {'q':'python'}    # q 는 키값, python은 value, https://www.google.co.kr/search?q=python
response = requests.get(URL,params=parameters)          # params 이용
print(response.status_code)
```

```python
print(response.text)
```

```python
URL = 'https://finance.naver.com/'
header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.72 Safari/537.36'}
parameters = {'query':'005930'}    # q 는 키값, python은 value, https://www.google.co.kr/search?q=python
response = requests.get(URL,headers=header, params=parameters)          # params 이용
print(response.status_code)
```

```python
print(response.text)
```
